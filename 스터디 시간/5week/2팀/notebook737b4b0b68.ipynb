make_subplots
>서브플롯 생성하기
* rows와 cols 값을 통해서 몇 행, 몇 열로 그래프를 그릴 지 설정
* column_titles: 열의 제목
* x_title: x(전체)의 제목

add_trace
> 이미 생성된 Figure 위에 덧붙여서 새로운 그래프를 그릴 때 사용
* 밑에 숫자는 각각 행과 열을 통한 위치를 말한다.

go.Bar
* orientation="h"는 수평으로 막대를 나타낸다
* marker는 그래프 디자인
    * color는 색깔의 범위
    * line_color는 테두리의 색깔
    * line_width는 선의 두께
    * coloraxis는 색깔의 범위를 그라데이션으로
    
update_layout
> 그래프 생성 완료 후 layout 정보를 업데이트 해주는 함수
* showlegend는 데이터의 종류를 표시하기 위한 텍스트, True하면 옆에 trace 별로(그래프별로) 나타낼 수 있다.
* title_text: 전체적인 제목
* title_x : 0이면 왼쪽 1이면 오른쪽 


```python
train_null=pd.DataFrame(train.isna().sum())
train_null=train_null.sort_values(by=0,ascending=False)[:-1]
test_null=pd.DataFrame(test.isna().sum())
test_null=test_null.sort_values(by=0,ascending=False)

fig=make_subplots(rows=1, cols=2, column_titles=["Train Data","Test Data"], x_title="Missing Values")

fig.add_trace(go.Bar(x=train_null[0],
                    y=train_null.index,
                    orientation="h",
                    marker=dict(color=[n for n in range(12)],
                               line_color='rgb(0,0,0)',
                               line_width=2,
                               coloraxis="coloraxis")),
             1,1)
fig.add_trace(go.Bar(x=test_null[0],
                    y=test_null.index,
                    orientation="h",
                    marker=dict(color=[n for n in range(12)],
                               line_color='rgb(0,0,0)',
                               line_width=2,
                               coloraxis="coloraxis")),
             1,2)

fig.update_layout(showlegend=False, title_text="Column wise Null Value Distribution", title_x=0.5)
```
