# 5. Data Pre-Processing

누락된 값 대치

```python
imputer_cols = ["Age", "FoodCourt", "ShoppingMall", "Spa", "VRDeck", "RoomService"]
imputer = SimpleImputer(strategy = STRATEGY) # 결측값 중간값으로 채우는 설정
imputer.fit(train[imputer_cols]) # 중간값 계산
train[imputer_cols] = imputer.transform(train[imputer_cols]) # 변형된 특성 있는 배열 넣음
test[imputer_cols] = imputer.transform(test[imputer_cols])
train["HomePlanet"].fillna('Z', inplace=True) # NaN값 Z로 채우기
test["HomePlanet"].fillna('Z', inplace=True)
```

### 범주형 인코딩

```python
label_cols = ["HomePlanet", "CryoSleep", "Cabin", "Destination", "VIP"]
def label_encoder(train, test, columns):
    for col in columns:
        train[col] = train[col].astype(str) # 문자형으로 변경
        test[col] = test[col].astype(str)
        train[col] = LabelEncoder().fit_transform(train[col])
        test[col] = LabelEncoder().fit_transform(test[col])
    return train, test

train, tset = label_encoder(train, test, label_cols)
```

```python
train.drop(["Name", "Cabin"], axis = 1, inplace = True) # 버리기
test.drop(["Name", "Cabin"], axis = 1, inplace = True)
X = train.drop(TARGET, axis = 1)
y = train[TARGET]
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 12, test_size = 0.33)
```

# 6. Modeling

27가지 분류기

## LGBM 분류

```python
lgb_params = { # 학습할 데이터 받기
    'objective' : 'binary',
    'n_estimators' : 50,
    'learning_rate' : 0.08
}

lgb_predictions = 0
lgb_scores = []
lgb_fimp = []
LGBM_FEATURES = list(train.columns)[:-1]
skf = StratifiedKFold(n_splits=FOLDS, shuffle=True, random_state = RANDOM_STATE) # 교차검증
for fold, (train_idx, valid_idx) in enumerate(skf.split(train[LGBM_FEATURES], train[TARGET])):
    print(f'\033[94m')
    print(10*"=", f"Fold={fold+1}", 10*"=")
    start_time = time.time() # 시간 측정
          
    X_train, X_valid = train.iloc[train_idx][LGBM_FEATURES], train.iloc[valid_idx][LGBM_FEATURES]
    y_train, y_valid = train[TARGET].iloc[train_idx], train[TARGET].iloc[valid_idx]
    
    model = LGBMClassifier(**lgb_params) # 데이터 불러와 LGBM 분류기로 모델 지정
    model.fit(X_train, y_train, verbose = 0) # 모델 학습
          
    preds_valid = model.predict(X_valid) # 모델 예측
    acc = accuracy_score(y_valid, preds_valid)
    lgb_scores.append(acc)
    run_time = time.time() - start_time
          
    fim = pd.DataFrame(index = LGBM_FEATURES,
                      data = model.feature_importances_,
                      columns = [f'{fold}_importance'])
    lgb_fimp.append(fim)
          
    print(f"Fold={fold+1}, Accuracy score: {acc:.2f}%, Run Time: {run_time:.2f}s")
    test_preds = model.predict(test[LGBM_FEATURES])
    lgb_predictions += test_preds/FOLDS
          
print("")
print("Mean Accuracy :", np.mean(lgb_scores))
```

## 항목 중요도

```python
lgbm_fis_df = pd.concat(lgb_fimp, axis = 1).head(15) # 동일한 형태의 DataFrame 합치기
lgbm_fis_df.sort_values('1_importance').plot(kind = 'barh', figsize = (15, 10), # 값 기준 정렬, 그래프 생성
                                            title = 'Feature Importance Across Folds')
plt.show() # 그래프 보기
```
