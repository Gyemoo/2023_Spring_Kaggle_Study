# 4. 학습된 모델을 사용하여 기계 학습 모델 및 예측을 구축한다.

```python
#importing all the required ML packages
from sklearn.ensemble import RandomForestClassifier # 유명한 randomforestclassfier 입니다. 
from sklearn import metrics # 모델의 평가를 위해서 씁니다
from sklearn.model_selection import train_test_split # traning set을 쉽게 나눠주는 함수입니다.
```

**Sklearn**
- feature engineering, preprocessing, 지도학습 알고리즘, 비지도 학습 알고리즘, 모델 평가, 파이프라인 등 머신러닝에 관련된 모든 작업이 인터페이스로 구현. 따라서 필수적으로 숙지.
- 현재 타이타닉 문제는 target class(survived) 이며, 이는 0, 1로 이루어진 binary classfication 문제라는 것을 알 수 있음.
- 모델을 최적화 시켜 테스트 데이터 셋의 생존 유무를 예측

```python
X_train = df_train.drop('Survived', axis=1).values
target_label = df_train['Survived'].values
X_test = df_test.values
```

```python
X_tr, X_vld, y_tr, y_vld = train_test_split(X_train, target_label, test_size=0.3, random_state=2018)
```

- train과 test도 중요하지만 이를 검증하는 것도 중요.
- random_state : 호출할 때마다 동일한 학습/테스트용 데이터 세트를 생성하기 위해 주어지는 난수 값.
- train_test_split은 랜덤으로 데이터를 분리하므로 설정하지 않으면 수행할 때마다 다른 학습/테스트 데이터 set가 생성. 따라서 random_state를 통해 수행 시 결과 값을 동일하게 맞춰준다.
- random_state에는 어떤 숫자를 적든 상관 없다고 말하지만, seed에 따른 구성이 어떻게 돼있는지는 중요하지 않은걸까..

```python
model = RandomForestClassifier()
model.fit(X_tr, y_tr)
prediction = model.predict(X_vld)
```

- sklearn에서 많은 머신러닝 알고리즘을 지원해준다길래 사이트에 들어가봐서 확인해보니 103개 정도가 존재하고 있었다.
- 현재 랜덤포레스트 모델을 사용하고 있다.
- 랜덤 포레스트는 결정트리기반 모델이며, 여러 결정 트리들을 앙상블한 모델.
- 머신러닝 알고리즘에는 여러 파라미터들이 있는데, 랜덤 포레스트분류기도 n_estimators(생성할 트리 개수), max_features(최대 선택할  특성의 수), max_depth(트리의 최대 깊이), min_samples_split(노드를 분할하기 위한 최소한의 샘플 데이터수), min_samples_leaf(리프노드가 되기 위해 필요한 최소한의 샘플 데이터수) 등 여러 파라미터들이 존재, 이들이 어떻게 세팅 되는지에 따라 모델 성능 다름.

```python
print('총 {}명 중 {:.2f}% 정확도로 생존을 맞춤'.format(y_vld.shape[0], 100 * metrics.accuracy_score(prediction, y_vld)))
```

**다음을 통해 파라미터들이 어떻게 세팅되는지에 따라 모델 성능이 달라질 수 있다는 것을 확인.**
- model = RandomForestClassifier(n_estimators = 10, max_features = 5 ) 로 할 경우 정확도 83.21%
- model = RandomForestClassifier() 로 할 경우 정확도 82.84%

# Feature importance
- 학습된 모델은 feature importance를 가진다고 한다. 따라서 어떤 feature에 영향을 많이 받았는지 확인이 가능하다.

```python
from pandas import Series

feature_importance = model.feature_importances_
Series_feat_imp = Series(feature_importance, index=df_test.columns)
```

- pandas series를 이용하면 sorting 하여 그래프를 그릴 수 있다.

```python
plt.figure(figsize=(8, 8))
Series_feat_imp.sort_values(ascending=True).plot.barh()
plt.xlabel('Feature importance')
plt.ylabel('Feature')
plt.show()
```

- feature importance 는 지금 모델에서의 importance이기 때문에 다른 모델을 사용하게 된다면 다르게 나올 수 있다.
- Fare가 중요한 특징 이라고 판단할 수 있지만, 이는 모델에 귀속되는 결론이기에 더 살펴봐야한다.
- 이를 통해 feature importance가 낮은 데이터 셋을 삭제하거나 정확도 높은 model을 얻기 위해 feature selection을 할 수 있다.
